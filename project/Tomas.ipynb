```python
import requests
import base64

CLIENT_ID = "client_marketdata_api"
CLIENT_SECRET = "client_marketdata_api"
SCOPE = "marketdata_api"
TOKEN_URL = "https://sts.nordpoolgroup.com/connect/token"

USERNAME = "s252876@student.dtu.dk"
PASSWORD = "gotqig-kyNja3-wynjiz"

credentials = f"{CLIENT_ID}:{CLIENT_SECRET}"
encoded = base64.b64encode(credentials.encode()).decode()

headers = {
    "Authorization": f"Basic {encoded}",
    "Content-Type": "application/x-www-form-urlencoded",
}

body = {
    "grant_type": "password",
    "scope": SCOPE,
    "username": USERNAME,
    "password": PASSWORD,
}

response = requests.post(TOKEN_URL, headers=headers, data=body)
print(response.text)  # check for errors
response.raise_for_status()

access_token = response.json()["access_token"]
print("Token retrieved successfully!")

body = {
    "grant_type": "password",
    "scope": SCOPE,
    "username": USERNAME,
    "password": PASSWORD,
}

# --- Request token ---
response = requests.post(TOKEN_URL, headers=headers, data=body)
response.raise_for_status()

token_data = response.json()
access_token = token_data["access_token"]
expires_in = token_data["expires_in"]  # seconds

print(f"Token: {access_token}")
print(f"Expires in: {expires_in}s")
```

```python
import pandas as pd

# --- Call API ---
params = {
    "market": "DayAhead",
    "areas": "EE,LT,LV,FI",
    "currency": "EUR",
    "date": "2024-01-01",
}

response = requests.get(
    "https://data-api.nordpoolgroup.com/api/v2/Auction/Prices/ByAreas",
    headers={"Authorization": f"Bearer {access_token}"},
    params=params,
)
response.raise_for_status()

# --- Parse into DataFrame ---
rows = []
for area_data in response.json():
    area = area_data["deliveryArea"]
    for price in area_data["prices"]:
        rows.append({
            "area": area,
            "deliveryStart": price["deliveryStart"],
            "deliveryEnd": price["deliveryEnd"],
            "price": price["price"],
        })

df = pd.DataFrame(rows)
df["deliveryStart"] = pd.to_datetime(df["deliveryStart"])
df["deliveryEnd"] = pd.to_datetime(df["deliveryEnd"])

print(df.head())

```
